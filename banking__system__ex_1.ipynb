```python
balance=1000
while True:
     print("""
     1. check balance
     2. withdrawn
     3.deposit
     4.exit""")
     choice=input("enter choice(1,2,3,0r4):")
     if choice=="1":
        print(f"your balance is:{balance}")
     elif choice=="2":
        amount=int(input("how much to withdrawn?"))
        if amount>balance:
           print("not enough money")
        else:
          balance-=amount
          print(f"new balance:{balance}")
     elif choice=="3":
         amount=int(input("how much to deposit?"))
         print(f"new balance{balance}")
     elif choice=="4":
         print("exiting good bye")
         break
     else:
        print("invalid choice")
```
