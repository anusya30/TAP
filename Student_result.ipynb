```python
# Base Class
class Student:
    def __init__(self, name, rollNumber):
        self.name = name
        self.rollNumber = rollNumber

    def display(self):
        print("Name:", self.name)
        print("Roll Number:", self.rollNumber)


# Derived Class - UndergraduateStudent
class UndergraduateStudent(Student):
    def __init__(self, name, rollNumber, m1, m2, m3):
        super().__init__(name, rollNumber)
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    def calculateAverage(self):
        avg = (self.m1 + self.m2 + self.m3) / 3
        print("Average Marks:", avg)


# Derived Class - PostgraduateStudent
class PostgraduateStudent(Student):
    def __init__(self, name, rollNumber, m1, m2, m3, m4):
        super().__init__(name, rollNumber)
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3
        self.m4 = m4

    def calculateAverage(self):
        avg = (self.m1 + self.m2 + self.m3 + self.m4) / 4
        print("Average Marks:", avg)


# Object Creation
ug = UndergraduateStudent("Anusya", 101, 80, 85, 90)
pg = PostgraduateStudent("Anu", 201, 75, 80, 85, 90)

print("---- Undergraduate Student ----")
ug.display()
ug.calculateAverage()

print("\n---- Postgraduate Student ----")
pg.display()
pg.calculateAverage()
```
